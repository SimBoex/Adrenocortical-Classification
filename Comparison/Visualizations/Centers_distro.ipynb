```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data from the image
data = {
    "Center 1":   [281, 122, 22, 45, 48, 19],
    "Center 2":   [14, 4, 0, 0, 0, 0],
    "Center 3":   [61, 11, 1, 0, 0, 0],
    "Center 4":   [39, 4, 0, 0, 0, 0],
    "Center 5":   [47, 14, 0, 0, 0, 0],
    "Center 6":   [19, 17, 8, 0, 0, 0],
    "Center 7":   [0, 0, 9, 0, 0, 0],
    "Center 8":   [18, 3, 0, 0, 0, 0],
    "Center 9":   [34, 14, 0, 0, 0, 0],
    "Center 10":  [3, 2, 2, 0, 0, 0],
    "Center 11":  [25, 11, 38, 15, 0, 0],
    "Center 12":  [32, 15, 4, 0, 1, 0],
    "Center 13":  [17, 11, 4, 0, 0, 0],
    "Center 14":  [0, 0, 12, 0, 0, 0]
}
classes = ["LRA", "BIAM", "ACC", "Pheochromocytoma", "Metastases", "Endothelial Adrenal Tumours"]

# Create DataFrame
df = pd.DataFrame(data, index=classes)

# Set the style and color palette
sns.set_theme(style="whitegrid")
plt.figure(figsize=(14, 6))

# Create the heatmap
ax = sns.heatmap(df, annot=True, fmt="d", cmap="YlOrBr", cbar_kws={"label": "Count"}, linewidths=0.5)

# Access the colorbar and set padding
cbar = ax.collections[0].colorbar
cbar.ax.set_ylabel("Count", labelpad=20, fontsize = 15)  # Adjust pad as needed
# Improve layout
plt.title("Class Distribution Across Centers", fontsize=16, pad=12)
plt.xlabel("Center", fontsize=15, labelpad=20)
plt.ylabel("Class", fontsize=15, labelpad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()

plt.savefig("/home/simone/Scrivania/AdreCorPredictions/Comparison/Visualizations/heatmap_class_vs_centers.png", dpi=300)
plt.show()
```
