# Metrics across 10 splits

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

scores_rus = pd.read_csv('../../Elements/RusBoost/AcrossSplits/scores.csv')
scores_boruta = pd.read_csv('../../Elements/Boruta_RusBoost/AcrossSplits/scores.csv')
scores_Multi = pd.read_csv('../../Elements/Multi_SMOTE_LogReg/AcrossSplits/scores.csv')
scores_LogReg = pd.read_csv('../../Elements/SMOTE_LogReg/AcrossSplits/scores.csv')

data = [
    scores_boruta["F1-score"].iloc[0:5],
    scores_rus["F1-score"].iloc[0:5],
    scores_Multi["F1-score"].iloc[0:5],
    scores_LogReg["F1-score"].iloc[0:5]
]

# Set the figure size
plt.figure(figsize=(12, 8))

# Create a bar chart with custom colors
colors = ['#FFA500', '#FF8C00', '#D8BFD8', '#9370DB']
box = plt.boxplot(data, patch_artist=True, notch=False, vert=True, labels=["Boruta_RusBoost", "RusBoost", "Multi_SMOTE_LogReg", "SMOTE_LogReg"])

# Customizing the box plot
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# Customizing other aspects of the box plot (median, whiskers, caps)
plt.setp(box['medians'], color='black', linewidth=2)
plt.setp(box['whiskers'], color='black', linewidth=1.5)
plt.setp(box['caps'], color='black', linewidth=1.5)

# Add labels and title
plt.xlabel('Models', fontsize=14)
plt.ylabel('F1-scores', fontsize=14)
plt.title('F1-score Distributions Across 5 Outer Folds for Each Model', fontsize=16, pad=60)

# Set y-axis limits and grid
plt.ylim(0.6, 1)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Optional: Remove the upper and right spines for a cleaner look
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Show the plot
plt.tight_layout()
plt.show()
```
