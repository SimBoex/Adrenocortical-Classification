```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
def FoundFeatures(Path):
    # reading the found dataset for the training
    df = pd.read_excel(Path)
    Features = set(df.columns)
    return Features
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Defining the model
model = "Boruta_RusBoost"

# Defining the dictionary containing the features
diz_Boruta_RusBoost = {}
for split in range(5):
    features_at_split = FoundFeatures("../../Elements/Boruta_RusBoost/AcrossSplits/" + "Split{}_Boruta_RusBoost_ReduceDataset.xlsx".format(split))
    diz_Boruta_RusBoost[split] = features_at_split

# Defining the matrix
intersection_matrix_Boruta_RusBoost = np.zeros((5, 5), dtype=float)

# Calculate the intersection matrix
for i in range(5):
    for j in range(i + 1, 5):
        value = len(diz_Boruta_RusBoost[i].intersection(diz_Boruta_RusBoost[j]))
        A = len(diz_Boruta_RusBoost[i])
        B = len(diz_Boruta_RusBoost[j])
        minumum = min(A, B)
        intersection_matrix_Boruta_RusBoost[i, j] = value / minumum

# Get the number of features for each split
feature_counts = [len(diz_Boruta_RusBoost[split]) for split in range(5)]

# Create a figure with two subplots: one for the heatmap and one for the bar plot
fig, ax = plt.subplots(2, 1, figsize=(6, 8), gridspec_kw={'height_ratios': [6, 1]})

# Plot the heatmap
ax[0].imshow(intersection_matrix_Boruta_RusBoost, cmap='Blues', interpolation='nearest')
ax[0].set_title('Features intersection across splits\n'+model)
ax[0].set_xticks(np.arange(len(intersection_matrix_Boruta_RusBoost)))
ax[0].set_yticks(np.arange(len(intersection_matrix_Boruta_RusBoost)))
ax[0].set_xlabel('Split')
ax[0].set_ylabel('Split')

# Adding text annotations for better clarity
for (i, j), val in np.ndenumerate(intersection_matrix_Boruta_RusBoost):
    if val != 0:  # Only annotate non-zero elements
        ax[0].text(j, i, f'{round(val, 1)}', ha='center', va='center', color='black')

# Plot the bar plot for the number of features
ax[1].bar(np.arange(5), feature_counts, color='#003366')
ax[1].set_xticks(np.arange(5))
ax[1].set_xlabel('Split')
ax[1].set_ylabel('Number of Features')
ax[1].set_title('Number of Features for outer fold')

# Increase the y-axis limit
ax[1].set_ylim(0, max(feature_counts) + 15)

# Add counts on top of each bar
for i, count in enumerate(feature_counts):
    ax[1].text(i, count + 0.5, str(count), ha='center', va='bottom', fontsize=10)


# Adjust layout
plt.tight_layout()
plt.show()
```
