```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV,RepeatedStratifiedKFold, cross_validate

# Creating a pandas dataframe of the patients
db = pd.read_excel("../Dataset/IntegratingSecretion_Cleaned_dataset.xlsx")

# removing the id and the subtypes of a given dataset
def extract_PATIENT_INFO(db):
   ids =  db.pop("ID")
   morf_codificata = db.pop("morf_codificata")
   return db, ids, morf_codificata

# extrating the label
y = db.pop("maligno")
X = db
# Splitting
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42,stratify=y)
```

```python
Train = pd.concat([X_train, y_train ], axis=1)
Train.head()
```

```python
Train.to_excel("../SNF_AdrenocorticalTumors/Train_Dataset.xlsx", index=False)
```
